```python

# tokenization
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt_tab')

text = "I loved the movie! It was absolutely amazing."
tokens = word_tokenize(text)
print(tokens)
```

```python
#removing stopwords

from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

filtered_tokens = [
    word for word in tokens if word.lower() not in stop_words
]

print(filtered_tokens)
```

```python
# stemming

from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

stemmed_tokens = [
    stemmer.stem(word) for word in filtered_tokens
]

print(stemmed_tokens)
```

Build a TF-IDF text classifier for sentiment usid IMDb data


```python
from datasets import load_dataset

dataset = load_dataset("imdb")

texts = dataset["train"]["text"]
labels = dataset["train"]["label"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words="english")
X = vectorizer.fit_transform(texts)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X, labels)
```

```python
from sklearn.metrics import accuracy_score, classification_report

predictions = model.predict(X)

print("Accuracy:", accuracy_score(labels, predictions))
print(classification_report(labels, predictions))
```

```python
#reviewing new text
review = ["The movie was boring and slow"]

review_tfidf = vectorizer.transform(review)
prediction = model.predict(review_tfidf)

print("Positive" if prediction[0] == 1 else "Negative")
```
